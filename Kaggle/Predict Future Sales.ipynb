```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('./datasets/competitive-data-science-predict-future-sales/sales_train.csv')
```

```python
dataset.head(5)
```

```python
sns.lineplot(data=dataset, x='date',y='item_price')
```

```python
temp_data = dataset.groupby(['date','shop_id','item_id']).sum().reset_index()
```

```python
temp_data
```

```python
sns.lineplot(data=temp_data, x='date',y='item_price')
```

```python
temp_data['MA'] = temp_data['item_price'].rolling(window=7).mean()
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=temp_data, x='date',y='MA')
```
